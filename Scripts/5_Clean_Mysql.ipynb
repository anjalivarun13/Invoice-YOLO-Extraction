```python
df['shipping_address']=df['billing_address']
```

```python
df.isnull().sum()
```

```python
# Strip spaces from column names
df.columns = df.columns.str.strip()
```

```python
# Clean invoice_id → keep only digits
df['invoice_id'] = df['invoice_id'].str.extract(r'(\d+)')
```

```python
# Clean customer_id → extract only CUST-### pattern
df['customer_id'] = df['customer_id'].str.extract(r'(CUST-\d+)')
```

```python
# Clean total column → remove S or any other unwanted characters
df['amount'] = df['amount'].astype(str).str.replace(r'[^0-9.]', '', regex=True)
df['subtotal'] = df['subtotal'].astype(str).str.replace(r'[^0-9.]', '', regex=True)
df['tax_due'] = df['tax_due'].astype(str).str.replace(r'[^0-9.]', '', regex=True)
df['total'] = df['total'].astype(str).str.replace(r'[^0-9.]', '', regex=True)
```

```python
df.tail(5)
```

```python
df.info()
```

```python
df['invoice_id'] = df['invoice_id'].astype(int)        # Convert to integer                   
```

```python
# Convert numeric columns
numeric_cols = ['amount', 'subtotal', 'tax_due', 'total']
df[numeric_cols] = df[numeric_cols].apply(pd.to_numeric, errors='coerce')

# Convert date columns
df['date'] = pd.to_datetime(df['date'], errors='coerce')
df['due_date'] = pd.to_datetime(df['due_date'], errors='coerce')

# Verify
print(df.dtypes)
```

```python
df.isnull().sum()
```

```python
# fillling null values
# Usually due_date is 1 month after date, so you can fill nulls automatically:

df['due_date'] = df['due_date'].fillna(df['date'] + pd.DateOffset(months=1))
```

```python
df.isnull().sum()
```

```python
# now we have null only in amount '29 null' 
# filling amount column null values with it's mean

df['amount'] = df['amount'].fillna(df['amount'].mean())
```

```python
df.isnull().sum()
```

```python
# now we don't have any null value our file is cleaned now.

# Save cleaned DataFrame to CSV
df.to_csv('cleaned_invoices.csv', index=False)
```
